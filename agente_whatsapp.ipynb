# Helisky Charter — FAQ no WhatsApp Web

Chatbot por regras (palavras-chave) para uma empresa fictícia de charter de helicóptero. O `faq_agente.responder()` é o “cérebro”; o resto é Selenium abrindo o WhatsApp Web e lendo a última mensagem recebida.

Código no pacote **`whatsapp_bot/`**: `cli.py` (entrada), `config.py` (`ALVOS_CHATS`), `session.py`, `sidebar.py`, `composer.py`, `messages.py` (leitura), `loop.py`, `faq_agente.py`.

Rodar: `uv run python -m whatsapp_bot` ou `uv run python main.py`.

Notebook de referência (FAQ por regras): [exemplo no Colab](https://colab.research.google.com/drive/1nF_5EoNisjk8kLdiwZjZI8iwnEw6mrnM?usp=sharing).

## 1. Testar o FAQ sem abrir o WhatsApp

```python
casos = [
    "Oi, bom dia",
    "Quero alugar um helicóptero",
    "Quanto custa o passeio?",
    "Posso pilotar o helicóptero?",
    "Quantas pessoas cabem?",
    "Que documentos preciso levar?",
    "É seguro voar?",
    "Quanto tempo dura o passeio panorâmico?",
    "E se estiver chovendo?",
    "Quero cancelar minha reserva",
    "Qual o telefone da Helisky?",
    "Obrigado!",
    "Vocês fazem entrega de pizza de helicóptero?",
]

for pergunta in casos:
    print("P:", pergunta)
    print("R:", responder(pergunta))
    print("-" * 60)
```
